```python
import pandas as pd,numpy as np, seaborn as sns, matplotlib.pyplot as plt
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.head(10)
```

```python
color=['#ADA1F0','#A1F0CB','#F0ACA1','#F0E284','#9E7D78']
sns.set_palette(sns.color_palette(color))
```

```python
sns.scatterplot(data=mpg,x='displ',y='hwy',hue='drv');
```

```python
sns.scatterplot(data=mpg, x='cty',y='hwy',hue='drv');
```

```python
ec=pd.read_csv('economics.csv')
ec.head()
```

```python
ec['date2']=pd.to_datetime(ec['date'])
ec['year']=ec['date2'].dt.year
ec['month']=ec['date2'].dt.month
sns.lineplot(data=ec, x = 'year',y='unemploy',errorbar=None);
```

```python
sns.lineplot(data=ec,x='year',y='psavert',errorbar=None);
```

```python
month_2014=ec.query('year==2014')
sns.lineplot(data=month_2014,x='month',y='psavert',errorbar=None);
```
